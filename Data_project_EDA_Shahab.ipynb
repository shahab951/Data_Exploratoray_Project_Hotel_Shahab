```python
#Bootstrapping and showing normal distribution for two lists of "stays_in_week_nights" and "stays_in_weekend_nights"
lm_a=[]
lm_b=[]
list_a=df['stays_in_week_nights']
list_b=df['stays_in_weekend_nights']
for i in range(10000):
    sample_a=np.random.choice(list_a,len(list_a))
    sample_b=np.random.choice(list_b,len(list_b))
    mean_a=np.mean(sample_a)
    mean_b=np.mean(sample_b)
    lm_a.append(mean_a)
    lm_b.append(mean_b)
```

```python
plt.hist(lm_a,bins=30)
plt.xlabel('Average of staying_in_week_nights')
plt.ylabel('Dsitribution')
plt.show()
```

```python
plt.hist(lm_b,bins=30)
plt.xlabel('Average of staying_in_weekend_nights')
plt.ylabel('Dsitribution')
plt.show()
```

```python
#Emprical cumilative distribution function
x=np.sort(list_a)
y=np.arange(1,len(x)+1)/len(x)
_=plt.plot(x,y, marker='.', linestyle='none')
_=plt.xlabel('Percentage of staying_in_week_nights')
_=plt.ylabel('ECDF')
plt.margins(0.02)
plt.show()
```

```python
x=np.sort(list_b)
y=np.arange(1,len(x)+1)/len(x)
_=plt.plot(x,y, marker='.', linestyle='none')
_=plt.xlabel('Percentage of staying_in_weekend_nights')
_=plt.ylabel('ECDF')
plt.margins(0.02)
plt.show()
```

```python
Diff_observed=np.mean(list_a)-np.mean(list_b)
```

```python
#Test statistics and null hypothesis
bs_replicates=[]
for i in range(10000):
    sample_a=np.random.choice(list_a,len(list_a))
    sample_b=np.random.choice(list_b,len(list_b))
    sample_difference=np.mean(sample_a)-np.mean(sample_b)
    bs_replicates.append(sample_difference)
```

```python
bs_replicates=np.array(bs_replicates)
```

```python
p=np.sum(bs_replicates>=Diff_observed)/10000
print(p)
```
